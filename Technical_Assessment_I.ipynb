```python
print("// ARITHMETIC SEQUENCE WITH N NUMBERS //") #title of program
a = 2 #initial number
N= int(input('N: ')) #input
#execute the program
if N == 0:
  print('The arithmetic sequence is:',0)
elif N < 0:
  print("Answer is not found. Please input N >= 0")
elif N > 0:
  print('The arithmetic sequence is:',str(a), end='')
  for i in range(N-1):
    a+=3
    print(","+ str(a), end='')
```
